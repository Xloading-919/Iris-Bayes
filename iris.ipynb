```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# 读取数据集
iris = pd.read_csv('D:\XData\jupyter note\Iris\iris.csv')
iris = np.array(iris)
print(iris)
# print(list(iris.columns))
```

```python
# 获取特征和标签
features = iris[:, :-1]
lable = iris[:, -1]
print(features)
print(lable)

# 划分数据集
x_train, x_test, y_train, y_test = train_test_split(features, lable, test_size=0.3)#, random_state=0)
```

```python
# 计算先验概率
pri_probs = {}
# 统计每个类别的样本个数
Species_count = {}
#得到类别数据
lables = set(y_train)
print(lables)

for l in lables:
    count = 0
    for s in y_train:
        if (s == l):
            count += 1
    Species_count[l] = count
    pri_prob = count / len(y_train)
    pri_probs[l] = pri_prob
print(Species_count)
print(pri_probs)
```

```python
# 预测类别
def predict(test_attr, pri_probs):

    # 密度估计(使用高斯密度)
    species = ''
    maxProb = 0

    #对每个类别的特征计算条件概率，最后计算后验概率
    for l in lables:
        pos_prob = pri_probs[l]#先验概率
        count = 0
        for attr in test_attr:
            #找到对应类别的每个特征的值
            attr_list = []  # 特征值列表
            index = np.argwhere(y_train == l)  # 记录每个类别对应特征的索引
            for i in index:
                attr_list.append(x_train[i, count])

            # 计算均值
            attr_mean = np.mean(attr_list)
            # 计算方差
            attr_var = np.std(attr_list)**2
            # 计算高斯密度得到类条件密度
            Gauss = np.exp((-1) * (attr-attr_mean)**2 / (2 * attr_var)) / np.sqrt(2 * np.pi * attr_var)

            #先验概率乘条件概率
            pos_prob *= Gauss
            count += 1
        #后验概率大的对应的类别作为预测结果
        if pos_prob > maxProb:
            maxProb = pos_prob
            species = l

    return species
```

```python
#测试结果
count = 0
for i in range(len(x_test)):
    pre = predict(x_test[i], pri_probs)
    if pre == y_test[i]:
        count += 1
    print("测试集第%d个标签为：%s，预测为：%s" % (i+1, y_test[i], pre))

print("准确率：%f" % (count/len(x_test)))
```
